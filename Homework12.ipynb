### Задание 3

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python

ratings['count'] = ratings.groupby('userId')['userId'].transform('count')
```

```python
def groupby_function(data):
    return data.timestamp.max() - data.timestamp.min()
```

```python
result = ratings.loc[ratings['count'] >100].groupby('userId').apply(groupby_function).mean()
```

```python
result
```
