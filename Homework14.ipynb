### Выбросы

```python

q1 = analyze_horse_3.pulse.quantile(0.25)
q3 = analyze_horse_3.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = analyze_horse_3[analyze_horse_3.pulse.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

```python
#мало что знаю о лошадях, но пульс 184 показался мне подозрительным, но, вероятно, такое вполне возможно, учитывая, что янво присутствует сильная боль и вздутие
pd.concat([analyze_horse_3, remove_outliers]).drop_duplicates(keep=False)
```

```python
# обращает на себя внимание размах дыхания от 8 до 96
q1 = analyze_horse_3['respiratory rate'].quantile(0.25)
q3 = analyze_horse_3['respiratory rate'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers2 = analyze_horse_3[analyze_horse_3['respiratory rate'].between(lower_bound, upper_bound, inclusive=True)].sort_values('respiratory rate')
remove_outliers2
```

```python
# видно, что случаев учащенного дыхания не так уж и мало и почти все они на фоне боли, учащенного пульса и повышенной температуры, а слишком медленное дыхание вообще не выходит за пределы нашей выборки
pd.concat([analyze_horse_3, remove_outliers2]).drop_duplicates(keep=False)
```
