### Задание 1

```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
def class_calc(param):
    if param <= 2:
        return "низкий рейтинг"
    elif param <= 4:
        return "средний рейтинг"
    elif param == 4.5 or param == 5:
        return "высокий рейтинг"         
```

```python
ratings['class']=ratings['rating'].apply(class_calc)
```

### Задание 2

```python
keywords =pd.read_csv('keywords.csv')
keywords.head()
```

```python
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
```

```python
def regions(k_word):
    for key,values in geo_data.items():
        for  value in values:
            if value in k_word.lower():
                return key
            else:
                return 'undefined'
```

```python
keywords['region']=keywords['keyword'].apply(regions)
keywords[keywords['region'] !='undefined']
```

### Задание 3

```python
years = list(range(1950,2011))
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
def production_year(title):
    for year in years:
        if str(year) in title:
            return str(year)
    return '1900'    
```

```python
movies['year']=movies['title'].apply(production_year)
```

```python
movies
```

```python
joined = ratings.merge(movies, how='left', on='movieId')
```

```python
joined[joined['year']!='1900'].groupby('year').rating.mean().sort_values(ascending=False)
```
