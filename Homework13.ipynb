```python
def get_all_links(url, query, pages):
    all_refs = {'date':[],'title':[],'link':[]}
    params = {
        'q': query
    }
    for i in range(pages):
        params['page'] = i + 1
        res = requests.get(url, params)
        time.sleep(0.3)
        soup = BeautifulSoup(res.text, 'html.parser')
        news_blocks = soup.find_all('h2', class_='post__title')
        a_list = list(map(lambda x: x.find('a',class_= 'post__title_link').get('href'), news_blocks))
        t_list = list(map(lambda x: x.find('a',class_= 'post__title_link').text, news_blocks))
        date = list(map(lambda x: x.find('span', class_='post__time').text, soup.find_all('header', class_='post__meta')))
        all_refs['date'] += date 
        all_refs['title']+= t_list
        all_refs['link'] += a_list
    return pd.DataFrame(all_refs)

all_links = get_all_links('https://habr.com/ru/search/', KEYWORDS, 1)
all_links
```

```python
df=[]
for link in all_links['link']:
    soup = BeautifulSoup(requests.get(link).text, 'html.parser')
    time.sleep(0.3)
    date = pd.to_datetime(soup.find('span', class_='post__time').get('data-time_published')).date()
    title =  soup.find('span', class_='post__title-text').text
    text =  soup.find('div', class_='post__text_v1').text
    df.append((date,title,link,text))

pd.DataFrame(df, columns=("date", "title", "link","text"))
```
